## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st
import os
from scipy.stats import linregress
from sklearn import datasets



# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
merged_data_df = pd.merge(mouse_metadata, study_results, how='left', on="Mouse ID")
# Display the data table for preview
merged_data_df.head(10)
```

```python
# Checking the number of mice.
number_of_mice = len(merged_data_df["Mouse ID"].unique())
#Show number of mice in summary table
number_of_mice_df = pd.DataFrame([{"Total Number of Mice": number_of_mice}])
number_of_mice_df
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
grouped_df = merged_data_df[['Mouse ID','Timepoint']]

# Optional: Get all the data for the duplicate mouse ID. 
duplicatemouse_df = grouped_df.loc[grouped_df.duplicated()]
duplicatemouse_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
everything_else_df = merged_data_df[merged_data_df["Mouse ID"].isin(duplicatemouse_df['Mouse ID'])==False]
everything_else_df
```

```python
# Checking the number of mice in the clean DataFrame.
clean_number_of_mice = len(everything_else_df["Mouse ID"].unique())
clean_number_of_mice
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
# and SEM of the tumor volume for each regimen
# This method is the most straighforward, creating multiple series and putting them
# all together at the end.
mean = everything_else_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean()
med = everything_else_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].median()
var = everything_else_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].var()
stdv = everything_else_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].std()
sem = everything_else_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem()
stat_summary_table = pd.DataFrame({"Mean": mean, "Median": med, "Variance": var, 
                                   "Standard Deviation": stdv, "SEM": sem
                                  })
stat_summary_table
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen
# This method produces everything in a single groupby function
tumor_volume_summary = everything_else_df.groupby("Drug Regimen").agg({"Tumor Volume (mm3)":["mean", "median", "var", "std", "sem"]})
tumor_volume_summary
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
# Start by getting the last (greatest) timepoint for each mouse
# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
#get greatest ("max") timepoint for each mouse
max = everything_else_df.groupby("Mouse ID")["Timepoint"].max()
max = pd.DataFrame(max)
#merge max dataframe wit clean dataframe to get tumor volume at the max timepoint
whygodwhy = max.merge(everything_else_df, on=["Mouse ID", "Timepoint"])
whygodwhy.head()
#view data    
Final_Tumor_Volume_Table = whygodwhy[["Drug Regimen", "Mouse ID", "Timepoint", "Tumor Volume (mm3)"]]
Final_Tumor_Volume_Table
```

```python
# Put treatments into a list for for loop (and later for plot labels)
# Create empty list to fill with tumor vol data (for plotting)
# Calculate the IQR and quantitatively determine if there are any potential outliers. 
    # Locate the rows which contain mice on each drug and get the tumor volumes
    # add subset 
    # Determine outliers using upper and lower bounds

#state four treatment regimens of interest
Treatment_Regimen = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
#group data by drug regimen and tumor volume
Tumor_Volume_Data = Final_Tumor_Volume_Table.groupby('Drug Regimen')['Tumor Volume (mm3)'].apply(list)
#make list into dataframe
Tumor_Volume_Data_df = pd.DataFrame(Tumor_Volume_Data)
#for-loop to establish IQR and possible outliers for each of the four specified regimens, print out same
volume_list = []
for drug in Treatment_Regimen:
    final_volume = Final_Tumor_Volume_Table.loc[Final_Tumor_Volume_Table["Drug Regimen"]==drug, "Tumor Volume (mm3)"] 
    volume_list.append(final_volume)
    quartiles = final_volume.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = round(upperq-lowerq,2)
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    print(f"The IQR for regimen at 'Treatment Regimen' List Item {drug} is {iqr}'.")
    if (lower_bound < final_volume.min()) & (upper_bound > final_volume.max()):
        print(f"\tThere are no outliers at 'Treatment Regimen' List Item {drug}'.")
    else:
        print(f"\tThere is at least 1 outlier at 'Treatment Regimen' List Item {drug}'.")
        


    

```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
plt.boxplot(volume_list, labels = Treatment_Regimen)
plt.title("Final Tumor Volume of Each Mouse for Specified Regimens")
plt.xlabel("Drug Regimen")
plt.ylabel("Tumor Volume, mm3")

plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin
#use previous clean dataframe to view only Capomulin Drug Regimen data
capomulin = everything_else_df.loc[everything_else_df["Drug Regimen"]=="Capomulin"]
#pick one mouse from said data (s185)
mouse_s185 = capomulin.loc[capomulin["Mouse ID"]=="s185"]
#generate, format, and show line plot of time point versus tumor volume for s185
plt.plot(mouse_s185["Timepoint"], mouse_s185["Tumor Volume (mm3)"], color="green")
plt.title("Timepoint vs Tumor Volume for Mouse s185 Treated with Capomulin")
plt.xlabel("Timepoint")
plt.ylabel("Tumor Volume, mm3")
plt.show()



```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen
#take capomulin data from previous step and group it by Mouse ID and establish the mean/average tumor volume
cap_avg = capomulin.groupby(["Mouse ID"]).mean()
#only show weight & tumor volume of mice
cap_avg = cap_avg[["Weight (g)","Tumor Volume (mm3)"]]
#format and show plot
cap_avg.plot(kind="scatter", x='Weight (g)', y='Tumor Volume (mm3)', color = "green", figsize=(8,8), title="Capomulin Regimen: Mouse Weight vs Tumor Volume")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

#name x and y data for slope
x_values = cap_avg['Weight (g)']
y_values = cap_avg['Tumor Volume (mm3)']
#run regression and store returned values
slope, intercept, rvalue, pvalue, stderr = linregress(x_values, y_values)
# Calculate the predictions for each value of x (Weight (g))
regression_values = x_values * slope + intercept
# store the string version of said linear equation, to be displayed on plot
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
#format and show plot with best fit line and slope showing
plt.figure(figsize=(10,8))
plt.scatter(x_values, y_values)
plt.plot(x_values,regression_values,"red")
plt.annotate(line_eq, (18,45), fontsize=15, color="red")
plt.title("Capomulin Regimen: Mouse Weight vs Tumor Volume")
plt.xlabel("Weight (g)")
plt.ylabel("Tumor Volume (mm3)")
print(f"The measure of the proportion of variability in the data, r^2 is: {rvalue**2}")
plt.show()

```
